```python
#Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

ANS- Python program that calculates and prints the sum and product of the first 10 natural numbers using both a for loop and a while loop.

python
# Using for loop
def sum_and_product_for():
    sum_n = 0
    product_n = 1
    for i in range(1, 11):
        sum_n += i
        product_n *= i
    return sum_n, product_n

# Using while loop
def sum_and_product_while():
    sum_n = 0
    product_n = 1
    i = 1
    while i <= 10:
        sum_n += i
        product_n *= i
        i += 1
    return sum_n, product_n

# Main code
if __name__ == "__main__":
    sum_for, product_for = sum_and_product_for()
    sum_while, product_while = sum_and_product_while()

    print("Using for loop:")
    print("Sum:", sum_for)
    print("Product:", product_for)

    print("\nUsing while loop:")
    print("Sum:", sum_while)
    print("Product:", product_while)
Explanation:
The sum_and_product_for function calculates the sum and product of the first 10 natural numbers using a for loop.
The sum_and_product_while function does the same using a while loop.
The main section prints the results for both methods.
```

```python
#Create a python program to compute the electricity bill for a household.The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 perunit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
You are required to take the units of electricity consumed in a month from the user as input.
Your program must pass this test case: when the unit of electricity consumed by the user in a month is
310, the total electricity bill should be 2250.


ANS- Python program to compute the electricity bill for a household based on the given per-unit charges. The program will take the number of units consumed in a month as input from the user and calculate the total bill accordingly.

python
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 4.5
    elif units <= 200:
        bill = 100 * 4.5 + (units - 100) * 6
    elif units <= 300:
        bill = 100 * 4.5 + 100 * 6 + (units - 200) * 10
    else:
        bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units - 300) * 20
    return bill

if __name__ == "__main__":
    try:
        # Take input from the user
        units_consumed = int(input("Enter the number of units consumed: "))

        # Calculate the bill
        bill_amount = calculate_electricity_bill(units_consumed)

        # Print the result
        print(f"Total Electricity Bill: Rs. {bill_amount:.2f}")
    except ValueError:
        print("Please enter a valid number for units consumed.")

Explanation:
Function Definition: The calculate_electricity_bill function calculates the bill based on the number of units consumed, using conditional statements to apply the appropriate rates.

User Input: The program prompts the user to enter the number of units consumed.

Error Handling: A try-except block is used to handle any invalid input, ensuring the program doesn't crash if the user enters a non-integer value.

Output: The total bill is printed, formatted to two decimal places.
```

```python
#Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list.

ANS- Python program that creates a list of numbers from 1 to 100, calculates the cube of each number, and appends the number to a list if its cube is divisible by 4 or 5. The program uses both a for loop and a while loop to demonstrate the calculation.

python
# Function to calculate cubes and filter based on divisibility using for loop
def calculate_cubes_for():
    divisible_numbers = []
    for i in range(1, 101):
        cube = i ** 3
        if cube % 4 == 0 or cube % 5 == 0:
            divisible_numbers.append(i)
    return divisible_numbers

# Function to calculate cubes and filter based on divisibility using while loop
def calculate_cubes_while():
    divisible_numbers = []
    i = 1
    while i <= 100:
        cube = i ** 3
        if cube % 4 == 0 or cube % 5 == 0:
            divisible_numbers.append(i)
        i += 1
    return divisible_numbers

# Main code
if __name__ == "__main__":
    # Using for loop
    divisible_numbers_for = calculate_cubes_for()
    print("Numbers whose cubes are divisible by 4 or 5 (for loop):", divisible_numbers_for)

    # Using while loop
    divisible_numbers_while = calculate_cubes_while()
    print("Numbers whose cubes are divisible by 4 or 5 (while loop):", divisible_numbers_while)


Explanation:
Function Definitions:

calculate_cubes_for: This function uses a for loop to iterate through numbers 1 to 100, calculates the cube of each number, and checks if it's divisible by 4 or 5. If so, it appends the number to a list.
calculate_cubes_while: This function achieves the same result using a while loop.
Main Code:

The program calls both functions and prints the results.
Output:
When you run the program, it will display the numbers whose cubes are divisible by 4 or 5 for both loop types. You can modify the range or conditions if needed!



```

```python
#Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"

ANS-Python program that counts the number of vowels in the given string:

python
# Given string
input_string = "I want to become a data scientist"

def count_vowels(s):
    vowels = "aeiouAEIOU"  # Including both lowercase and uppercase vowels
    count = 0

    for char in s:
        if char in vowels:
            count += 1

    return count

if __name__ == "__main__":
    vowel_count = count_vowels(input_string)
    print(f"Number of vowels in the string: {vowel_count}")


Explanation:
Input String: The string "I want to become a data scientist" is defined.
Function Definition: The count_vowels function:
Initializes a string of vowels (both lowercase and uppercase).
Iterates through each character in the input string and checks if it's a vowel.
Increments the count for each vowel found.
Main Code: The program calls the function and prints the total number of vowels.
Output:
When you run this program, it will output the number of vowels in the given string. For the provided string, the output will be:

Number of vowels in the string: 10
```
